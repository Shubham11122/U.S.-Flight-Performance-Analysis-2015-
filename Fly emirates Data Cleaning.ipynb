```python
remaining_cols = [
    'taxi_in',
    'wheels_on',
    'wheels_off',
    'taxi_out',
    'departure_time',
    'departure_delay',
    'tail_number',
    'scheduled_time'
]

for col in remaining_cols:
    print(f"\n----- {col} -----")
    print(
        flights_clean[
            flights_clean[col].isnull()
        ][['cancelled', 'diverted']]
        .value_counts()
        .head()
    )
```

```python
#labelling null with unknown in tail number because it indicates aircraft not assigned.
flights_clean['tail_number'] = flights_clean[
    'tail_number'
].fillna('Unknown')
```

```python
flights_clean['scheduled_time'].isnull().sum()
```

only 6 value are null in scheduled time which is 0.0001% that why I remove them. 

```python
flights_clean = flights_clean.dropna(
    subset=['scheduled_time']
)
```

```python
flights_clean['scheduled_time'].isnull().sum()
```

```python
flights_clean.isnull().sum().sort_values(ascending=False)
```

Missing values were investigated individually. Delay-cause columns (weather_delay, airline_delay, security_delay, air_system_delay, late_aircraft_delay) were filled with 0 because NULL represented the absence of that delay type. Cancellation_reason NULLs were replaced with "Not Cancelled". Tail_number NULLs were replaced with "Unknown". Operational flight columns such as departure_time, arrival_time, air_time, and elapsed_time were retained as NULL because they corresponded to cancelled or diverted flights and represent valid business scenarios.
